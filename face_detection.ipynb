```
# load the pre-trained model
from cv2 import CascadeClassifier
classifier = CascadeClassifier('haarcascade_frontalface_default.xml')
classifier
```

```
# load the photograph
from cv2 import imread
pixels = imread('test4.jpg')
```

```
# perform face detection
bboxes = classifier.detectMultiScale(pixels,1.05,8)
# (image, scaleFactor, minNeighbors)
# scale factor default value ->1.10. 1.05,1.4..
# min neighbors default ->3, 6,8...
```

```
# print bounding box for each detected face
for box in bboxes:
  print(box)
```

```
from cv2 import rectangle
# extract
for box in bboxes:
  x, y, width, height = box
  x2, y2 = x + width, y + height
# draw a rectangle over the pixels
  rectangle(pixels, (x, y), (x2, y2), (0,0,255), 2)
```
